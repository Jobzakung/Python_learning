```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
import ipywidgets as widgets
%matplotlib inline
```

```python
img = cv.imread('./imageEx/eight.tif', 0)
plt.imshow(img, cmap='gray')
```

```python

plt.hist(img.ravel(), 256, [0, 256])
plt.show()
```

```python
@widgets.interact(x = (0, 255), y = ["orginal image", "result image"])
def show(x ,y):
    ret, thresh3 = cv.threshold(img, x, 255, cv.THRESH_TRUNC)
    result = {"orginal image": img, "result image": thresh3}
    plt.imshow(result[y], cmap='gray')
```
