```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(6.4*5, 4.8*5), constrained_layout=False)

img_c1 = cv2.imread("period_input.jpg", 0)
H = np.ones(img_c1.shape[:2])
H[106-15:116+15,315-15:325+15]=0
H[166-15:176+15,235-15:245+15]=0
H[166-15:176+15,398-15:408+15]=0
H[230-15:240+15,150-15:160+15]=0
H[230-15:240+15,480-15:490+15]=0
H[290-15:300+15,235-15:245+15]=0
H[290-15:300+15,398-15:408+15]=0
H[350-15:360+15,315-15:325+15]=0
img_c2 = np.fft.fft2(img_c1)
img_c3 = np.fft.fftshift(img_c2)
img_c6 = H*np.fft.fftshift(img_c2)
img_c4 = np.fft.ifftshift(img_c6)
img_c5 = np.fft.ifft2(img_c4)
import matplotlib.pyplot as plt
plt.figure(figsize=(12.8*10, 9.6*5), constrained_layout=False)
plt.subplot(131), plt.imshow(np.log(1+np.abs(img_c6)), "gray"), plt.title("H*Centered Spectrum")
plt.subplot(132), plt.imshow(np.log(1+np.abs(img_c3)), "gray"), plt.title("Decentralized")
plt.subplot(133), plt.imshow(np.abs(img_c5), "gray"), plt.title("Processed Image")

plt.show()
```
