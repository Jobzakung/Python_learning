```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('ex4.png')

plt.imshow(image)
```

```python
kernel=np.ones((5,1),np.uint8)
horizontalLine = cv2.dilate(image,kernel,iterations=1)
horizontalLine = cv2.erode(horizontalLine,kernel,iterations=1)
```

```python
plt.imshow(horizontalLine)
```
